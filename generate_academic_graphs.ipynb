# Generate graphs from {t, python_nbs}

```python
py = python_nbs[0]
content = nb2content[py]
```

```python
import re
```

```python
def get_cells_from_content(content, file):
    code_cells = []
    temp_cell = []
    
    if "cells" in content:
        origin_cells = content["cells"]
    elif len(content["worksheets"])==1:
        origin_cells = content["worksheets"][0]["cells"]

            
    for c in origin_cells:
        temp_cell.append(c)
        if c["cell_type"]=='code':
            code_cells.append(temp_cell)
            temp_cell = []
        
        
    assert len(code_cells)==len([c for c in origin_cells if c["cell_type"]=='code'])
    cells = []
    prev_len=0
    for i, c in enumerate(code_cells):
        position=i
        target_lineno = prev_len
        if "source" in c[-1]:
            code = c[-1]["source"]
        else:
            code = c[-1]["input"]
        annotation = []
        for mc in c[:-1][::-1]:
            annotation+=mc["source"][::-1]
        annotation = [re.sub('[^0-9a-zA-Z]+', ' ', l) for l in annotation]
        prev_len+=len(code)
        cells.append({"position":position,
                     "target_lineno":target_lineno,
                      "file":file,
                     "code":code,
                     "annotation":annotation})
#         print()
#         raise NotImplementedError
    return cells
```

```python
cells = get_cells_from_content(content, py)
```

# Post Process graphs

```python
graphs = []
with open('./graphs/aca_graphs.txt','r') as f:
    for l in f:
        graphs.append(json.loads(l))
```

```python

for i,g in enumerate(graphs):
    g["id"]=i
    g["stage"]='0'
    g["header"]=""
    g["neighbor_cells"] = [-1]
```

```python
with open('./graphs/temp_aca_graphs.txt','w') as fout:
    for g in graphs:
        fout.write(json.dumps(g))
        fout.write('\n')
#         graphs.append(json.loads(l))
```
